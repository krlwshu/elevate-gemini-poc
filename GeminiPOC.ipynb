```python
from google import genai
import time
import json
import os
import dotenv

dotenv.load_dotenv()
API_KEY=your_api_key

client = genai.Client(api_key=API_KEY)

questions = [
    "Tell me about yourself.",
    "Why are you interested in this role?",
    "Describe a challenge you've overcome.",
]

def process_with_gemini(answer, context="You're an AI interviewer. Give constructive feedback."):
    try:
        prompt = f"{context}\n\nCandidate's answer:\n{answer}"
        response = client.models.generate_content(
            model="gemini-1.5-pro",
            contents=prompt
        )
        return response.text
    except Exception as e:
        return f"[Gemini error: {e}]"

# === üöÄ INTERVIEW FLOW ===
def run_interview():
    session = {
        "start_time": time.strftime("%Y-%m-%d %H:%M:%S"),
        "questions": []
    }

    print("üé§ Welcome to the Gemini Interview Flow!\n")

    for idx, question in enumerate(questions):
        print(f"üü° Q{idx + 1}: {question}")
        answer = input("üìù Your answer: ")

        feedback = process_with_gemini(answer)

        print(f"\nüß† Gemini Feedback:\n{feedback}\n{'-' * 40}\n")

        session["questions"].append({
            "question": question,
            "answer": answer,
            "gemini_feedback": feedback
        })

    session["end_time"] = time.strftime("%Y-%m-%d %H:%M:%S")

    with open("interview_session_gemini.json", "w") as f:
        json.dump(session, f, indent=2)

    print("‚úÖ Session saved to interview_session_gemini.json")

if __name__ == "__main__":
    run_interview()
```
